**ENUNCIADO:**

Rafa, Pedro y María quieren comprar palmeras, donuts, hojaldres y tartas. Cada uno necesita diferentes cantidades y pueden elegir dos tiendas: Manolo Bakes y Starbucks.

Las siguiente tablas muestran, respectivamente, la cantidad que cada uno necesita, y los precios en cada una de las tiendas:

image.png

**SE PIDE:** Genera la matriz que exprese la cantidad que cada persona gastaría en cada tienda, y recomienda a cada uno dónde debería hacer sus compras.


**Objetivo:** Aplicar la teoría de matrices a un caso práctico real.

### Importamos la librería numpy:

```python
import numpy as np
```

### Creamos las matrices con numpy:

```python
A = np.array([[6, 5, 3, 1],
              [3, 6, 2, 2],
              [3, 4, 3, 1]])

B = np.array([[1.50, 1.00],
              [2.00, 2.50],
              [5.00, 4.50],
              [16.00, 17.00]])
```

### Multiplicamos A x B

```python
resultado = A@B
resultado
```

Importamos Pandas para crear el Dataframe con las recomendaciones:

```python
import pandas as pd
```

```python
df = pd.DataFrame(resultado, index=['Rafa', 'Pedro', 'María'], columns=['Manolo Bakes', 'Starbucks'])
df
```

```python
for persona in df.index:
    fila = df.loc[persona]    # los gastos de esa persona
    tienda = fila.idxmin()    # devuelve la etiqueta o índice donde está el valor más bajo
    gasto = fila.min()        # devuelve el valor más bajo

    print(f"{persona}: tu compra sale más barata en {tienda} ({gasto} €)")
```

### Conclusiones y recomendaciones

A partir de la matriz generada con la operación `A@B` donde se indica cuánto gastaría cada persona en cada tienda, la recomendación para cada cliente es la tienda con menor gasto en su fila:

- Recomendaría a **Rafa ir a Starbucks** porque se ahorra 1€

- Recomendaría a **Pedro ir a Manolo Bakes** porque ahorra 2.5€. Lo que marca aquí la diferencia es que Pedro compra 2 tartas y en Starbucks son más caras, por eso le resulta más barato ir a Manolo Bakes.

- En cuánto a **María, gasta lo mismo en las dos tiendas**. Le recomendaría ir a Manolo Bakes por un tema de calidad (pastelería casera frente a bollería industrial siempre es más recomendable) o conveniencia (alterar cantidades en la compra para abaratarla).
